# Calculate the error rate between the actual and measured distance between bolt holes

```python
actual = pd.read_csv("./actual.csv", header=None)
```

```python
actual
```

```python
actual.drop(columns = [24], inplace=True)
```

```python
actual.dropna(axis=0, inplace=True)
```

## For axis = 0

```python
measured_x = pd.read_csv("./distances_x.csv", index_col=[0])
```

```python
measured_x
```

```python
measured_x = measured_x[measured_x != 0]
```

```python
measured_x.dropna(inplace=True)
```

```python
measured_x
```

### By median

```python
measured_x_m = measured_x.median(axis=0)
```

```python
measured_x_m
```

```python
actual_x_s = actual.iloc[0].squeeze()
```

```python
actual_x_s
```

```python
rows = 3

for i in range(0, rows):
    print(abs(actual_x_s.values - measured_x_m[i * 24:(i + 1) * 24].values) / actual_x_s.values * 100)
```

### By optimum

```python
rows = 3

df_info = pd.DataFrame()

for j in range(0, len(measured_x)):
    np_measured = np.array([])
    measured_x_s = measured_x.iloc[j].squeeze()

    for i in range(0, rows):
        np_measured = np.append(np_measured, abs((actual_x_s.values + 7) - (measured_x_s[i * 24:(i + 1) * 24].values + 7)) / (actual_x_s.values + 7) * 100)

    df = pd.DataFrame(np_measured)
    
    median = df.median()
    average = df.mean()

    df_info = pd.concat([df_info, pd.DataFrame([median, average])], ignore_index=True)

    # df.to_csv(f"{j} row.csv", index = False)
    # df_info.to_csv(f"{j} row_info.csv", index=False)
```

```python
df_info
```

```python
df_info.idxmin()
```

```python
df_info.iloc[df_info.idxmin() // 2]
```

```python
measured_x_o = measured_x.iloc[df_info.idxmin() // 2].squeeze()
```

```python
measured_x_o
```

```python
rows = 3

np_measured_absolute_x = np.array([])
np_measured_relative_x = np.array([])

for i in range(0, rows):
    np_measured_absolute_x = np.append(np_measured_absolute_x, abs((actual_x_s.values + 7) - (measured_x_s[i * 24:(i + 1) * 24].values + 7)))
    np_measured_relative_x = np.append(np_measured_relative_x, abs(((actual_x_s.values + 7) - (measured_x_o[i * 24:(i + 1) * 24].values + 7)) / (actual_x_s.values + 7)) * 100)

print(np_measured_absolute_x)
print(np_measured_relative_x)

df_x_absolute = pd.DataFrame(np_measured_absolute_x)
df_x_relative = pd.DataFrame(np_measured_relative_x)

absolute_x_median = df_x_absolute.median()
absolute_x_average = df_x_absolute.mean()

print(absolute_x_median)
print(absolute_x_average)

df_x_absolute_error_info = pd.DataFrame([absolute_x_median, absolute_x_average])
df_x_absolute_error_info.to_csv("absolute_error_x_info.csv", index=False)

relative_x_median = df_x_relative.median()
relative_x_average = df_x_relative.mean()

print(relative_x_median)
print(relative_x_average)

df_x_relative_error_info = pd.DataFrame([relative_x_median, relative_x_average])
df_x_relative_error_info.to_csv("relative_error_x_info.csv", index=False)
```

```python
np_measured_absolute_x.min()
```
