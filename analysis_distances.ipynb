```python
import cv2
```

```python
ref = pd.read_csv("./file.csv", header=None)
coordinates = pd.read_csv("./found_coordinates.csv")
```

```python
ref.rename(columns={0: 'centerX'}, inplace=True)
ref.rename(columns={1: 'centerY'}, inplace=True)
ref.rename(columns={2: 'radius'}, inplace=True)

bolt_holes_info = ref
```

```python
radius = bolt_holes_info['radius'].mean()
```

```python
import numpy as np
```

```python
img = np.zeros((2000, 2000, 3), np.uint8)
```

```python
for i in range(len(bolt_holes_info)):
        centerX = int(bolt_holes_info['centerX'].iloc[i])
        centerY = int(bolt_holes_info['centerY'].iloc[i])
        radius = int(radius)

        img = cv2.circle(img, (centerX, centerY), radius, (255, 255, 255), 3)
```

```python
for i in range(0, len(coordinates)):
    x1 = coordinates['found_x1'].iloc[i].item()
    y1 = coordinates['found_y1'].iloc[i].item()

    x2 = coordinates['found_x2'].iloc[i].item()
    y2 = coordinates['found_y2'].iloc[i].item()

    img = cv2.putText(img, str(round(df.median(axis=0).iloc[i].item(), 2)), (int((x1 + x2) / 2 - 15), int((y1 + y2) / 2 - 15)), cv2.FONT_HERSHEY_PLAIN, 1, (255, 255, 255), 1)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img)
plt.xlim([600, 2000])
plt.ylim([350, 600])
plt.show()
```

```python
cv2.imwrite("./output.png", img)
```

# Calculate the error rate between the actual and measured distance between bolt holes

```python
actual = pd.read_csv("./actual.csv", header=None)
```

```python
actual
```

```python
actual.drop(columns = [24], inplace=True)
```

```python
actual.dropna(axis=0, inplace=True)
```

```python
measured = pd.read_csv("./distances.csv", index_col=[0])
```

```python
measured
```

```python
measured = measured[measured != 0]
```

```python
measured.dropna(inplace=True)
```

```python
measured
```

```python
measured_m = measured.median(axis=0)
```

```python
measured_m
```

```python
actual_s = actual.squeeze()
```

```python
actual_s
```

```python
rows = 3

for i in range(0, rows):
    print(abs(actual_s.values - measured_m[i * 24:(i + 1) * 24].values) / actual_s.values * 100)
```
